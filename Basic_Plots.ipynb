# Scatter plots

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# random.random() generates numbers between 0.1 and 1.0
# random.random(50) generates 50 numbers between 0.1 and 1.0

x_data = np.random.random(50)*100  #range multiplied by 100: 0.1 to 1 becomes 1 to 100
y_data = np.random.random(50)*100  

plt.scatter(x_data, y_data)  #plt.scatter() plots the scatter plot of x_data and y_data
plt.show()
```

```python
# random.random() generates numbers between 0.1 and 1.0
# random.random(50) generates 50 numbers between 0.1 and 1.0

x_data = np.random.random(50)*100  #range multiplied by 100: 0.1 to 1 becomes 1 to 100
y_data = np.random.random(50)*100  

print(x_data)
print(y_data)
plt.scatter(x_data, y_data, c='red')  #color for points: c means color


plt.xlabel("x-axis")
plt.ylabel('y-axis')
plt.show()
```

# Line plots

```python
"""
year = []
for x in range(9):
    year = year + [1998 + x] 
    
print(year)

"""
year = [2009 + x for x in range(9)]
print(year)

weight = [80 - x for x in range(9)]  
print(weight)

#NOTE: x(year) and y(weight) must have the same dimension
plt.plot(year, weight, c='red', lw='3')  # color = red, 'lw' means line width
```

```python
year = [2009 + x for x in range(9)]
print(year)

weight = [90,123,105,92,185,182,194,193,23]  
print(weight)

# NOTE: x(year) and y(weight) must have the same dimension

plt.plot(year, weight, c='red', lw='3')  # color = red, 'lw' means line width
```

```python
year = [2009 + x for x in range(9)]
print(year)

weight = [90,123,105,92,185,182,194,193,23]  
print(weight)

# NOTE: x(year) and y(weight) must have the same dimension

plot1 = plt.plot(year, weight, c='red', lw='3', linestyle = '--')  # changing line style

# valid value for ls; supported values are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
```

# Bar plot

```python
x = ['python', 'java', 'c++', 'html', 'javascripts']
y = [200, 150, 400, 304, 283]

plt.bar(x, y) # bar graph
```

```python
x = ['python', 'java', 'c++', 'html', 'javascripts']
y = [200, 150, 400, 304, 283]

plt.bar(x, y, color = 'red' ) # NOTE bar graphs, you need to use 'color' instead of 'c'
```

```python
x = ['python', 'java', 'c++', 'html', 'javascripts']
y = [200, 150, 400, 304, 283]

plt.bar(x, y, color = 'red', align = 'edge' ) 
# align the graph's starting point with the first x-input "python"
```

```python
x = ['python', 'java', 'c++', 'html', 'javascripts']
y = [200, 150, 400, 304, 283]

plt.bar(x, y, color = 'red', align = 'edge', width = 0.1) 
# align the graph's starting point with the first x-input "python"
# bar width = 0.1
```

```python
x = ['python', 'java', 'c++', 'html', 'javascripts']
y = [200, 150, 400, 304, 283]

plt.bar(x, y, color = 'red', align = 'edge', width = 0.8, edgecolor = 'green', lw=6 ) 
# align the graph's starting point with the first x-input "python"
# bar width = 0.1
# edges of bars have a color green and a line width of 6
```
