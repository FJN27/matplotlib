# Plotting Functions
1. define domain with np.linspace
2. define function 
3. graphing: labeling axes 
4. using matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-3, 3, 1000)  #domain of x: [-3:3]
```

```python
def f(x):
    return x**2
```

```python
f(2)
```

```python
plt.plot(x, f(x))
```

```python
def f(x):
    return np.sin(x)  #np.sin(x) is the sine function. "sin" isnot defined 
```

```python
plt.plot(x, f(x))
```

```python
plt.plot(x, f(x))
plt.title('f(x)=sin(x)')  #title of the graph
plt.xlabel("x")   #title for x-axis
plt.ylabel("y")   # title for y-axis
```

```python
plt.plot(x, f(x))
plt.title('f(x)=sin(x)')  #title of the graph
plt.xlabel("x")   #title for x-axis
plt.ylabel("y")   # title for y-axis
plt.axhline(color = 'black')  # plot the x-axis at the origin. "axh" means a horizontal line
plt.axvline(color = 'black')  # plot the t-axis at the origin. "axv" means a vertical line
```

# plotting multiples functions on one graph

```python
def f(x):
    return np.sin(x)
def g(x):
    return np.cos(x)
plt.plot(x, f(x))
plt.plot(x, g(x))
```
